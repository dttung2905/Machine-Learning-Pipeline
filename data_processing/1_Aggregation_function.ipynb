```python
import pandas as pd
import numpy as np
```

```python
#import data

df = pd.read_csv('../dataset/Auto.csv')
df.head()
```

```python
def create_new_col(aggs):
    return [k + "_"+ v for k in aggs for v in aggs[k]]
```

```python
aggs = {'displacement':['mean','std'],
       'weight':['max','min']}

df_agg = df.groupby('origin').agg(aggs)
df_agg.columns = create_new_col(aggs)
df_agg.reset_index(inplace = True)
df_agg.head()
```
